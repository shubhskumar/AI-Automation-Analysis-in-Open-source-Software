```python
contents_df = spark.read.parquet("gs://msca-bdp-data-open/final_project_git/contents")
```

```python
selected_columns = ["content"]
contents_df_selected = contents_df.select(*selected_columns)
contents_df_selected = contents_df_selected.na.drop(subset=["content"])
contents_df_selected.show(5)
del contents_df
```

```python
from pyspark.sql.functions import col, expr, posexplode, split
import re
```

```python
data_science_libraries = ["pandas", "numpy", "matplotlib", "seaborn", "scipy", "scikit-learn", "tensorflow", "pytorch", "keras", "nltk"]

data_science_libraries_lower = [lib.lower() for lib in data_science_libraries]
```

```python
extracted_words_df = (
    contents_df_selected
    .select(posexplode(split(col("content"), " ")).alias("pos", "word"))
    .filter((col("word") != "") & (col("pos") < 20))
    .drop("pos")
)
```

```python
library_counts_df = (
    extracted_words_df
    .filter(col("word").isin(data_science_libraries_lower))
    .groupBy("word")
    .count()
    .orderBy(col("count").desc())
)
```
